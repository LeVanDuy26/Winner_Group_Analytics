```python
# ==========================
# 1. Import & Kết nối DB
# ==========================
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv

# Load biến môi trường
load_dotenv()

DB_USER   = os.getenv("DB_USER")
DB_PASS   = os.getenv("DB_PASS")
DB_HOST   = os.getenv("DB_HOST")
DB_PORT   = os.getenv("DB_PORT")
DB_SILVER = os.getenv("DB_SILVER")
DB_GOLD   = os.getenv("DB_GOLD")

engine_silver = create_engine(
    f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_SILVER}"
)
engine_gold = create_engine(
    f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_GOLD}"
)
```

```python
# Load dim_shop
df_shop = pd.read_sql("SELECT * FROM dim_shops", engine_silver)
print("dim_shop:", df_shop.shape)
```

```python
# Load vào Gold
df_shop.to_sql(
    "gold_dim_shops",
    engine_gold,
    if_exists="replace",
    index=False
)

print("✅ Đã load gold_dim_shops vào schema Gold.")
```
